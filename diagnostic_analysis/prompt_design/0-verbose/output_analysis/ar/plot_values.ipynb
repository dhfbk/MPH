```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
macro = [
    0.557,
    0.457,
    0.431,
    0.534,
    0.477,
    0.469,
    0.524,
    0.306
]

plt.figure(figsize=(20, 5))
sns.barplot(x=configurations, y=macro)
plt.axhline(np.mean(macro), color='black', linestyle='--')
plt.axhline(macro[0], color='black', linestyle='--')
plt.axhline(macro[-1], color='black', linestyle='--')
#set color of the bars to red if the value is below the mean, to green if it is above the mean and to blue if it is the value of FAKE CONV
colors = ['r' if v < np.mean(macro) else 'b' if v > np.mean(macro) else 'b' for v in macro]
for i, bar in enumerate(plt.gca().patches):
    bar.set_color(colors[i])
    


plt.show()
```
