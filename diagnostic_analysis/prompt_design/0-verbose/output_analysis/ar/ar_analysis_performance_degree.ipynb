```python
import pickle
import networkx as nx
from process_graphs import extract_graphs, find_all_undirected_graphlet_instances, get_instance_ratios, find_possible_n_nodes2

with open('data/medium_ubuntu.pkl', 'rb') as f:
    data = pickle.load(f)

```

```python
ar_conv_out_path = 'out_ar/ar_conv_out.pkl'
ar_conv_struct_path = 'out_ar/ar_conv_struct_out.pkl'
ar_struct_summ_path = 'out_ar/ar_struct_summ_out.pkl'
ar_struct_desc_path = 'out_ar/ar_struct_desc_out.pkl'
ar_struct_summ_desc_path = 'out_ar/ar_struct_summ_desc_out.pkl'
```

```python
graph_set, indeces_connected_graphs = extract_graphs(data)

#save graphlets to file
import pickle

with open("tool_items/graphlets.pkl", "rb") as f:
    graphlet = pickle.load(f)

undirect_graphlet_instances = []
number_of_nodes = []
degree_node_0 = []

for g in graph_set:
    d, n = find_all_undirected_graphlet_instances(g, graphlet)
    #extract degree of node 0G.to_undirected()
    new_g = g.to_undirected()
    
    deg = new_g.degree(0)
    
    undirect_graphlet_instances.append(d)
    degree_node_0.append(deg)
    number_of_nodes.append(n)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

with open(ar_conv_out_path, 'rb') as f:
    ar_conv_out = pickle.load(f)
    
with open('data/medium_ubuntu_true_addresses.pkl', 'rb') as f:
    gold = pickle.load(f)
    
row_1 = [x for x,y in zip(ar_conv_out['row'], degree_node_0) if y == 1]
row_2 = [x for x,y in zip(ar_conv_out['row'], degree_node_0) if y == 2]
row_3 = [x for x,y in zip(ar_conv_out['row'], degree_node_0) if y == 3]

out_1 = [x for x,y in zip(ar_conv_out['output'], degree_node_0) if y == 1]
gold_1 = [x for x,y in zip(gold, degree_node_0) if y == 1]
out_2 = [x for x,y in zip(ar_conv_out['output'], degree_node_0) if y == 2]
gold_2 = [x for x,y in zip(gold, degree_node_0) if y == 2]
out_3 = [x for x,y in zip(ar_conv_out['output'], degree_node_0) if y == 3]
gold_3 = [x for x,y in zip(gold, degree_node_0) if y == 3]

print(sum(ar_conv_out['row'])/len(ar_conv_out['row']))
print("\n")

print(sum(row_1)/len(row_1))
print(sum(row_2)/len(row_2))
print(sum(row_3)/len(row_3))

cm = confusion_matrix(out_1, gold_1, labels=[0, 1, 2, 3, 4])

# Define column and row names
columns = ['0', '1', '2', '3', '4']
rows = ['0', '1', '2', '3', '4']

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16}, xticklabels=columns, yticklabels=rows)
plt.show()

cm = confusion_matrix(out_2, gold_2, labels=[0, 1, 2, 3, 4])

# Define column and row names
columns = ['0', '1', '2', '3', '4']
rows = ['0', '1', '2', '3', '4']

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16}, xticklabels=columns, yticklabels=rows)
plt.show()

cm = confusion_matrix(out_3, gold_3, labels=[0, 1, 2, 3, 4])

# Define column and row names
columns = ['0', '1', '2', '3', '4']
rows = ['0', '1', '2', '3', '4']

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16}, xticklabels=columns, yticklabels=rows)
plt.show()
```
