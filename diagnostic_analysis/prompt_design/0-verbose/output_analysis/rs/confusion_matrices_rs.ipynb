```python
rs_conv_path = 'out_rs/rs_conv_out.pkl'
rs_conv_struct_path = 'out_rs/rs_conv_struct_out.pkl'
rs_struct_summ_path = 'out_rs/rs_struct_summ_out.pkl'
rs_struct_desc_path = 'out_rs/rs_struct_desc_out.pkl'
rs_struct_summ_desc_path = 'out_rs/rs_struct_summ_desc_out.pkl'
```

```python
import pickle

with open(rs_conv_path, 'rb') as f:
    rs_conv_out = pickle.load(f)

with open(rs_conv_struct_path, 'rb') as f:
    rs_conv_struct_out = pickle.load(f)

with open(rs_struct_summ_path, 'rb') as f:
    rs_struct_summ_out = pickle.load(f)

with open(rs_struct_desc_path, 'rb') as f:
    rs_struct_desc_out = pickle.load(f)

with open(rs_struct_summ_desc_path, 'rb') as f:
    rs_struct_summ_desc_out = pickle.load(f)
```

```python
print(sum(rs_conv_out["row"])/len(rs_conv_out["row"]))
print(sum(rs_conv_struct_out["row"])/len(rs_conv_struct_out["row"]))
print(sum(rs_struct_summ_out["row"])/len(rs_struct_summ_out["row"]))
print(sum(rs_struct_desc_out["row"])/len(rs_struct_desc_out["row"]))
print(sum(rs_struct_summ_desc_out["row"])/len(rs_struct_summ_desc_out["row"]))
```

```python
#confusion matrix between rs_c and rs_s
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(rs_conv_out["output"], rs_conv_struct_out["output"])

print("CONV vs CONV+STRUCT")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()

cm = confusion_matrix(rs_conv_struct_out["output"], rs_struct_summ_out["output"])

print("CONV + STRUCT vs STRUCT + SUMM")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()


cm = confusion_matrix(rs_conv_struct_out["output"], rs_struct_desc_out["output"])

print("CONV + STRUCT vs STRUCT + DESC")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()
```

```python
cm = confusion_matrix(rs_struct_summ_out["output"], rs_struct_desc_out["output"])

print("STRUCT+SUMM vs STRUCT+DESC")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()

cm = confusion_matrix(rs_struct_summ_desc_out["output"], rs_struct_summ_out["output"])

print("STRUCT+SUMM+DESC vs STRUCT+SUMM")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()

cm = confusion_matrix(rs_struct_summ_desc_out["output"], rs_struct_desc_out["output"])

print("STRUCT+SUMM+DESC vs STRUCT+DESC")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()
```

```python
cm = confusion_matrix(rs_conv_struct_out["output"], rs_struct_summ_desc_out["output"])

print("CONV+STRUCT vs STRUCT+SUMM+DESC")
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 16})
plt.show()
```
