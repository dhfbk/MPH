```python
import pickle
from process_graphs import extract_graphs, find_all_undirected_graphlet_instances, get_instance_ratios, find_possible_n_nodes2

with open('data/medium_ubuntu.pkl', 'rb') as f:
    data = pickle.load(f)

```

```python
rs_conv_path = 'out_rs/rs_conv_out.pkl'
rs_conv_struct_path = 'out_rs/rs_conv_struct_out.pkl'
rs_struct_summ_path = 'out_rs/rs_struct_summ_out.pkl'
rs_struct_desc_path = 'out_rs/rs_struct_desc_out.pkl'
rs_struct_summ_desc_path = 'out_rs/rs_struct_summ_desc_out.pkl'
```

```python
graph_set, indeces_connected_graphs = extract_graphs(data)

#save graphlets to file
import pickle

with open("tool_items/graphlets.pkl", "rb") as f:
    graphlet = pickle.load(f)

undirect_graphlet_instances = []
number_of_nodes = []

for g in graph_set:
    d, n = find_all_undirected_graphlet_instances(g, graphlet)
    undirect_graphlet_instances.append(d)
    number_of_nodes.append(n)
```

```python
with open(rs_conv_path, 'rb') as f:
    rs_conv_out = pickle.load(f)

nn = number_of_nodes[0]

graph_right = []
for i in range(len(undirect_graphlet_instances[0])):
    max = find_possible_n_nodes2(4, nn[i])
    int_graph = []
    for j in range(int(max)+1):
        int_graph.append(0)
    graph_right.append(int_graph)

graph_tot = []
for i in range(len(undirect_graphlet_instances[0])):
    max = find_possible_n_nodes2(4, nn[i])
    int_graph = []
    for j in range(int(max)+1):
        int_graph.append(0)
    graph_tot.append(int_graph)

for i in range(len(undirect_graphlet_instances)):
    for j in range(len(undirect_graphlet_instances[i])):
        graph_tot[j][undirect_graphlet_instances[i][j]] += 1
        if rs_conv_out["row"][i] == 1:
            graph_right[j][undirect_graphlet_instances[i][j]] += 1



for gr, gt in zip(graph_right, graph_tot):
    gr.pop(0)
    gt.pop(0)

graph_ratios = []

for gr, gt in zip(graph_right, graph_tot):
    graph_ratios.append([x/y  if y != 0 else 0 for x, y in zip(gr, gt)])

import networkx as nx
import matplotlib.pyplot as plt

for g, gt, graph in zip(graph_ratios, graph_tot, graphlet):
    print(g)
    print(gt)
    #plot graph
    nx.draw(graph, with_labels=True, font_weight='bold')
    plt.show()
```

```python
for idx in range(len(graph_ratios)):
    if idx>1:
        for g_i in graph_ratios[idx]:
            print(g_i)
```
